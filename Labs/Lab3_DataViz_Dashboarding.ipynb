### Q1 ‚Äî Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

üìù **Your response:**

Exploratory Visualization:

Audience: Usually yourself or your immediate team.
Purpose: To understand the data, find patterns, identify outliers, and generate hypotheses. It's about discovery.
Level of Polish: Often rough, quick, and iterative. You're rapidly trying different views of the data.
Iteration: High. You'll likely create many visualizations, discarding most, as you home in on insights.

Explanatory Visualization:

Audience: Others (stakeholders, clients, the public) who need to understand a specific finding or story from the data.
Purpose: To communicate a clear, specific message or insight effectively and persuasively. It's about explanation.
Level of Polish: High. These visualizations are carefully designed, annotated, and cleaned up for clarity and impact.
Iteration: Lower, focused on refining the message and visual design for a specific audience and purpose.

### Q2 ‚Äî Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

üìù **Your response:**
1. **Bar:**
   - Question archetype: Comparison (comparing values across different categories)
   - Data shape: Categorical variable on one axis, numerical variable on the other.
   - Pitfalls: Can be misleading with too many categories, difficult to read if bars are too thin or too wide, inappropriate for showing trends over continuous data.
   - Alternatives: Column chart (vertical bars), Stacked bar chart, Grouped bar chart.
2. **Line:**
   - Question archetype: Trend (showing how a value changes over time or another continuous variable)
   - Data shape: Continuous variable (often time) on the x-axis, numerical variable on the y-axis.
   - Pitfalls: Can be misleading with too many lines, difficult to read if lines overlap, inappropriate for showing discrete data.
   - Alternatives: Area chart, Scatter plot (with connected points), Time series plot.
3. **Scatter:**
   - Question archetype: Relationship/Correlation (showing the relationship between two numerical variables)
   - Data shape: Two numerical variables. Each point represents an observation with values for both variables.
   - Pitfalls: Can be difficult to see patterns with too many points (overplotting), can imply causation where there is only correlation, difficult to read if points are not clearly distinguishable.
   - Alternatives: Bubble chart (adds a third numerical variable), Heatmap (for large datasets), Line plot (if there's a clear order to the points).
4. **Histogram:**
   - Question archetype: Distribution (showing the frequency distribution of a single numerical variable)
   - Data shape: Single numerical variable. Data is grouped into bins, and the height of each bar represents the frequency of data points in that bin.
   - Pitfalls: The choice of bin size can significantly affect the appearance of the distribution, inappropriate for categorical data.
   - Alternatives: Density plot, Box plot, Violin plot.
5. **Map:**
   - Question archetype: Geospatial (showing how a value varies across geographic regions)
   - Data shape: Geographical data (e.g., latitude/longitude, region names) and a numerical or categorical variable to be displayed on the map.
   - Pitfalls: Can be misleading if the size of the region is not proportional to the value being displayed (choropleth maps), can be difficult to read with too many data points or overlapping markers, requires specialized software or libraries.
   - Alternatives: Dot map, Symbol map, Cartogram.

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

```
Given `df_flights` with `cost_usd`, `revenue_usd`, and `casm`, produce a **scatter plot** of cost vs. revenue with points colored by `casm` (continuous colormap) and add a diagonal y=x reference line. Include labels and legend.
```


````python
# Prompt A: On-Time Rate by Carrier (Bar Chart)
on_time_rate = df_flights.groupby('carrier')['on_time'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=on_time_rate.index, y=on_time_rate.values, palette='viridis')

plt.title('On-Time Rate by Carrier', fontsize=16)
plt.xlabel('Carrier', fontsize=12)
plt.ylabel('On-Time Rate (%)', fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add percentage labels on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
````

````python
# Prompt B ‚Äî Average Arrival Delay over Time (Line Chart)
daily_avg_delay = df_flights.groupby('date')['arr_delay_min'].mean().reset_index()
daily_avg_delay['rolling_avg_7d'] = daily_avg_delay['arr_delay_min'].rolling(window=7).mean()

plt.figure(figsize=(14, 7))
sns.lineplot(data=daily_avg_delay, x='date', y='arr_delay_min', label='Daily Average Delay', color='skyblue')
sns.lineplot(data=daily_avg_delay, x='date', y='rolling_avg_7d', label='7-Day Rolling Average', color='coral')

plt.title('Average Arrival Delay Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Arrival Delay (Minutes)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Prompt C ‚Äî Delay Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df_flights['arr_delay_min'], bins=50, kde=True)

plt.title('Distribution of Arrival Delays', fontsize=16)
plt.xlabel('Arrival Delay (Minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.axvline(0, color='red', linestyle='--', label='0 Minute Delay')

# Calculate and annotate percent of flights within [-5, 15] minutes
on_time_range = df_flights[(df_flights['arr_delay_min'] >= -5) & (df_flights['arr_delay_min'] <= 15)]
percent_on_time_range = (len(on_time_range) / len(df_flights)) * 100

plt.annotate(f'{percent_on_time_range:.1f}% of flights within [-5, 15] min',
             xy=(15, plt.gca().get_ylim()[1] * 0.8), # Position annotation near the right side
             xytext=(20, plt.gca().get_ylim()[1] * 0.7), # Offset text from point
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
             fontsize=10)

plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Prompt D ‚Äî Load Factor by Route (Top 15) (Horizontal Bar)
route_load_factor = df_flights.groupby(['origin', 'dest'])['load_factor'].mean().sort_values(ascending=False).head(15)

plt.figure(figsize=(12, 8))

# Format the MultiIndex to a list of strings for the y-axis labels
route_labels = [f'{origin} - {dest}' for origin, dest in route_load_factor.index]

sns.barplot(x=route_load_factor.values, y=route_labels, palette='viridis')

plt.title('Top 15 Routes by Average Load Factor', fontsize=16)
plt.xlabel('Average Load Factor', fontsize=12)
plt.ylabel('Route (Origin - Dest)', fontsize=12)
plt.xlim(0.5, 1.0)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
````

````python
# Prompt E ‚Äî Cost vs. Revenue with CASM Coloring (Scatter)
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df_flights, x='cost_usd', y='revenue_usd', hue='casm', palette='viridis', alpha=0.6)

plt.title('Cost vs. Revenue Colored by CASM', fontsize=16)
plt.xlabel('Cost (USD)', fontsize=12)
plt.ylabel('Revenue (USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Add a diagonal y=x reference line
x_max = max(df_flights['cost_usd'].max(), df_flights['revenue_usd'].max())
plt.plot([0, x_max], [0, x_max], color='red', linestyle='--', label='Cost = Revenue')

plt.legend(title='CASM')
plt.tight_layout()
plt.show()
````
